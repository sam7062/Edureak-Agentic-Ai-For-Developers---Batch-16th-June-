```python
from agno.agent import Agent
from agno.models.openai import OpenAIChat
```

```python
from agno.agent import Agent
from agno.models.openai import OpenAIChat

agent = Agent(
    model=OpenAIChat(id="gpt-4o-mini"),

    name="Professor AI",

    description="""
    You are an experienced AI professor.
    Explain every concept using simple examples.
    """,

    instructions=[
        "Be friendly.",
        "Use bullet points.",
        "Give practical examples.",
        "Avoid jargon unless necessary."
    ]
)
```

```python
agent.print_response(
    "Explain what an AI Agent is in simple words.",
    stream=True
)
```

```python
from agno.agent import Agent
from agno.models.openai import OpenAIChat

agent = Agent(

    model=OpenAIChat(
        id="gpt-4o-mini"
    ),

    # Agent metadata
    name="AI Mentor",

    description="""
    An AI mentor helping students learn Generative AI.
    """,

    # Behavioral instructions
    instructions=[
        "Explain concepts step by step.",
        "Always provide examples.",
        "Keep responses beginner friendly.",
        "Summarize at the end."
    ],

    # Markdown output
    markdown=True
)
```

```python
agent.print_response(
    "Explain what an AI Agent is in simple words.",
    stream=True
)
```
